# Before-Class Preparation: AI Models, APIs, and Agents Lab

To make sure we can spend our class time building and experimenting rather than troubleshooting installations, please complete the following preparation before coming to class.

## 1. Install Ollama and Download a Local Model

We will use local Large Language Models (LLMs) during the lab.

### Install Ollama

Download and install Ollama from:

https://ollama.com/download

### Verify the Installation

Open a terminal (PowerShell on Windows) and run:

```bash
ollama --version
```

### Download a Small Model

Please download at least one of the following models:

```bash
ollama pull llama3.2
```

or

```bash
ollama pull qwen3:4b
```

The first model is smaller and recommended if your laptop has limited resources.

### Test the Model

Run:

```bash
ollama run llama3.2
```

Ask a simple question such as:

```text
What is machine learning?
```

If the model responds, your setup is ready.

## 2. Fork and Prepare the Lab Repository

We will work in a shared GitHub workflow during the class.

### Step 1 – Fork the Repository

Fork the lab repository to your own GitHub account:

```text
https://github.com/AramRstN/hf-api-agents-lab
```

### Step 2 – Create a Working Directory

Create a directory on your computer where you keep your programming projects.

Example:

```text
C:\Projects
```

or

```text
D:\Programming
```

### Step 3 – Clone Your Fork

Open a terminal and navigate to your projects directory:

```bash
cd path/to/your/projects
```

Clone your fork:

```bash
git clone <YOUR_FORK_URL>
```

Enter the project directory:

```bash
cd hf-api-agents-lab
```

Open the project in your preferred IDE (PyCharm or VS Code).

### Step 4 – Run the Setup Notebook

Open and execute:

```text
notebooks/00_setup.ipynb
```

Run all cells from top to bottom and make sure all checks pass successfully.
